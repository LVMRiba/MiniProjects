```python
# Vamos criar uma função para retornar os genêros
def retorna_itens(df, column_name):
    df[column_name] = df[column_name].str.upper().values
    temp = df[column_name].dropna()
    vetor = CountVectorizer(token_pattern = '(?u)\\b[\\w-]+\\b', analyzer = 'word').fit(temp)
    lista_unica =  vetor.get_feature_names()
    lista_unica = [item for item in lista_unica if len(item) > 1]
    return lista_unica

# Aplica a função
paises_unicos = retorna_itens(db_titles, 'production_countries')

# Visualiza
paises_unicos
```

```python
# Vamos criar uma função para retornar os genêros
def retorna_itens(df, column_name):
    df[column_name] = df[column_name].str.lower().values
    temp = df[column_name].dropna()
    vetor = CountVectorizer(token_pattern = '(?u)\\b[\\w-]+\\b', analyzer = 'word').fit(temp)
    lista_unica =  vetor.get_feature_names()
    lista_unica = [item for item in lista_unica if len(item) > 1]
    return lista_unica

# Aplica a função
generos_unicos = retorna_itens(db_titles, 'genres')

# Visualiza
generos_unicos
```
