## SQLite - Inserir Dados

```python
# Cria a consulta SQL
consulta1 = '''SELECT type, COUNT(*) AS COUNT FROM titles GROUP BY type'''

def insert_credits(df):
    for row in range(0, len(df)):
        if row%15000 == 0:
            print(row)
        consulta_insert_credits = f"""INSERT INTO Credits (
        person_id, 
        id, 
        name, 
        character, 
        role, 
        stream) 
        VALUES (
        '{df['person_id'][row]}', 
        '{df['id'][row]}', 
        '{df['name'][row]}', 
        '{df['character'][row]}', 
        '{df['role'][row]}', 
        '{df['stream'][row]}'
        )"""
        cur.execute(consulta_insert_credits)
    
def insert_titles(df):
    for row in range(0, len(df)):
        if row%1000 == 0:
            print(row)
        consulta_insert_titles = f"""INSERT INTO Titles (
        id, 
        title, 
        type, 
        description, 
        release_year, 
        age_certification, 
        runtime, 
        genres, 
        production_countries, 
        seasons, 
        imdb_id,
        imdb_score, 
        imdb_votes, 
        tmdb_popularity, 
        tmdb_score, 
        stream) 
        VALUES (
        '{df['id'][row]}', 
        '{df['title'][row]}', 
        '{df['type'][row]}', 
        '{df['description'][row]}', 
        {df['release_year'][row]}, 
        '{df['age_certification'][row]}', 
        {df['runtime'][row]},
        "{df['genres'][row]}", 
        "{df['production_countries'][row]}", 
        {df['seasons'][row]}, 
        '{df['imdb_id'][row]}', 
        {df['imdb_score'][row]}, 
        {df['imdb_votes'][row]}, 
        {df['tmdb_popularity'][row]}, 
        {df['tmdb_score'][row]}, 
        '{df['stream'][row]}'
        )"""
        cur.execute(consulta_insert_titles)
```

```python
genero_counts = []
```

```python
for genero in generos_unicos:

    command_09 = f""" SELECT COUNT(*) AS Count FROM titles WHERE genres LIKE '%{genero}%' AND type='MOVIE' """
    temporario_00 = pd.read_sql_query(command_09, conn)
    print(f'{genero.upper()}')
    display(temporario_00)
    genero_counts.append(temporario_00.values[0][0])
```

```python
genero_counts
```

```python
genero_avg_imdb = []
genero_min_imdb = []
genero_max_imdb = []
```

```python
for genero in generos_unicos:

    command_10 = f""" SELECT AVG(imdb_score), MIN(imdb_score), MAX(imdb_score) 
    FROM titles WHERE genres LIKE '%{genero}%' AND imdb_score > 0 AND type='MOVIE' """
    temporario_01 = pd.read_sql_query(command_10, conn)
    print(f'{genero.upper()}')
    display(temporario_01)
    genero_avg_imdb.append(temporario_01.values[0][0])
    genero_min_imdb.append(temporario_01.values[0][1])
    genero_max_imdb.append(temporario_01.values[0][2])
```

```python
genero_avg_tmdb = []
genero_min_tmdb = []
genero_max_tmdb = []
```

```python
for genero in generos_unicos:

    command_11 = f""" SELECT AVG(tmdb_score), MIN(tmdb_score), MAX(tmdb_score) 
    FROM titles WHERE genres LIKE '%{genero}%' AND tmdb_score > 0 AND type='MOVIE' """
    temporario_02 = pd.read_sql_query(command_11, conn)
    print(f'{genero.upper()}')
    display(temporario_02)
    genero_avg_tmdb.append(temporario_02.values[0][0])
    genero_min_tmdb.append(temporario_02.values[0][1])
    genero_max_tmdb.append(temporario_02.values[0][2])
```

```python
genero_dados = pd.DataFrame()
genero_dados['genre'] = generos_unicos
genero_dados['count'] = genero_counts
genero_dados['avg_imdb'] = genero_avg_imdb
genero_dados['min_imdb'] = genero_min_imdb
genero_dados['max_imdb'] = genero_max_imdb
genero_dados['avg_tmdb'] = genero_avg_tmdb
genero_dados['min_tmdb'] = genero_min_tmdb
genero_dados['max_tmdb'] = genero_max_tmdb
genero_dados
```

### 1) Criando um Wordcloud

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Libraries
from wordcloud import WordCloud

text=''

for genero in genero_dados['genre']:
    word = genero + ' '
    text += f"{word}*{genero_dados['count'][generos_unicos.index(genero)]} "
    word = ''
    
# Create the wordcloud object
wordcloud = WordCloud(width=480, height=480, margin=0, background_color="white").generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
